````python
import os
import pypdfium2 as pdfium
````

````python
current_dir_name = os.getcwd()
# pdf_file_path = os.path.join(current_dir_name, 'input', 'Core_v6.0.pdf')
pdf_file_path = os.path.join(current_dir_name, 'input', 'the-bluetooth-le-primer-v1.2.0.pdf')
pdf = pdfium.PdfDocument(pdf_file_path)

for page in pdf:
    textpage = page.get_textpage()
    text = textpage.get_text_range()
    print(text)
````
